```python
src_columns_names = data.columns.values.tolist()
re_columns_names = [
    'age',
    'how_long',
    'time_for_learn',
    'grade_exist_services',
    'amount_money',
    'email',
    'exact_sciences',
    'humanitarian_sciences',
    'chemestry_biology',
    'psychology',
    'creativity',
    'IT',
    'business',
    'english_or_other'
]

columns_dict = dict(zip(src_columns_names, re_columns_names))
columns_dict
```

```python
data.rename(columns = columns_dict, inplace=True)
data
```

```python
def sum_category(df, list_columns):
    result = []
    for el in list_columns:
        result.append(df[el].sum())
    result_df = pd.DataFrame(
        list(zip(list_columns, result)),
        columns =['Category', 'count']
    )
    return result_df

count_categories_df = sum_category(data, re_columns_names[6:])
count_categories_df
```

```python
# Лидер (Как видно - в порывах изучения первое место занимает IT)
count_categories_df[count_categories_df['count'] == count_categories_df['count'].max()]
```

```python
# У опрошенных явно отсутствует интерес к Точным наукам, психологии, 'творчеству', иностранным языкам
count_categories_df[count_categories_df['count'] == count_categories_df['count'].min()]
```
